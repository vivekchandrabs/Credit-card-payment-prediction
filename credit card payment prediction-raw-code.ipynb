### To predict credit card payment using logistic regression using the raw code.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
%matplotlib inline
```

### Reading the data from the excel file.

```python
data_set = pd.read_excel("default of credit card clients.xls",header = 1)
```

### Droping the 'default payment next month' because it is the target
### Droping the 'ID' as its just the row number.

```python
data =  data_set.drop(['default payment next month','ID'], axis = 1)
```

### Considering 'default payment next month' as the target

```python
target = data_set['default payment next month']
```

### Obtaining the correlation using spearman correlation function.

```python
correlation = data_set.corr('spearman')
```

### From this we get to know that 'PAY_0', 'PAY_2', 'PAY_3', 'PAY_4' are the most impacting features.

### Converting into a numpy array.

```python
x1 = np.array(data['PAY_0'])
x2 = np.array(data['PAY_2'])
x3 = np.array(data['PAY_3'])
x4 = np.array(data['PAY_4'])


y = np.array(target)
```

### Normalizing the data.

```python
x1 = (x1-x1.min())/(x1.max()-x1.min())
x2 = (x2-x2.min())/(x2.max()-x2.min())
x3 = (x3-x3.min())/(x3.max()-x3.min())
x4 = (x4-x4.min())/(x4.max()-x4.min())



```

### Spliting the data using train_test_split method.

```python
xt1, xs1, xt2,xs2,xt3,xs3,xt4,xs4 ,yt, ys = train_test_split(x1,x2,x3,x4,y,test_size =0.5)
```

```python
def hypo(a,b,c,d,q,x1,x2,x3,x4):

    return 1/(1+np.exp(-((a*x1) +(b*x2) + (c*x3)+ (d*x4)+ q)))
```

```python
def error(a,b,c,d,q,x1,x2,x3,x4,y):
    e1 = 0 
    m = len(x1)
    
    for i in range(m):
        e1 += (-(y[i]*np.log(hypo(a,b,c,d,q,x1[i],x2[i],x3[i],x4[i]))) - ((1-y[i]) * np.log(1-hypo(a,b,c,d,q,x1[i],x2[i],x3[i],x4[i]))))
    
    return (1/m) * e1
```

```python
def step_gradient(a,b,c,d,q,x1,x2,x3,x4, y, learning_rate):
    grad_a = 0 
    grad_b = 0
    grad_c = 0
    grad_d = 0
 

    grad_q = 0
    
    m = len(x1)
    
    for i in range(m):
        grad_a += (hypo(a,b,c,d,q,x1[i],x2[i],x3[i],x4[i]) - y[i]) * x1[i]
        grad_b +=  (hypo(a,b,c,d,q,x1[i],x2[i],x3[i],x4[i]) - y[i]) * x2[i]
        grad_c +=  (hypo(a,b,c,d,q,x1[i],x2[i],x3[i],x4[i]) - y[i]) * x3[i]
        grad_d += (hypo(a,b,c,d,q,x1[i],x2[i],x3[i],x4[i]) - y[i]) * x4[i]
      
        grad_q += (hypo(a,b,c,d,q,x1[i],x2[i],x3[i],x4[i]) - y[i])
        
    a= a - grad_a * learning_rate
    b= b - grad_b * learning_rate
    c= c - grad_c * learning_rate
    d= d - grad_d * learning_rate


    q = q - grad_q * learning_rate
    
    return a,b,c,d,q
```

```python
def descend(init_a,init_b,init_c,init_d,init_q,x1,x2,x3,x4,y,learning_rate,iterations):
    a = init_a
    b = init_b
    c = init_c
    d = init_d


    q = init_q
    
    for i in  range(iterations):
        e1 = error(a,b,c,d,q,x1,x2,x3,x4,y)
        if i%100 == 0:
            print(f"error: {e1}, a:{a}, b: {b}, c: {c}, d: {d},  q: {q}")
        a,b,c,d,q= step_gradient(a,b,c,d,q,x1,x2,x3,x4, y, learning_rate)
    return a,b,c,d,q
```

```python
final_a = 0
final_b = 0
final_c = 0
final_d = 0

final_q = 0
iterations = 1000
learning_rate = 0.0001
```

```python
for i in range(3):
    final_a ,  final_b, final_c, final_d, final_q = descend(final_a ,  final_b, final_c, final_d, final_q,xt1,xt2,xt3,xt4,yt,learning_rate,iterations)
```

```python
def accuracy(a,b,c,d,q,x1,x2,x3,x4,y):
    correct = 0
    m = len(x1)
    
    for i in range(m):
        pred = hypo(a,b,c,d,q,x1[i],x2[i],x3[i],x4[i])
        if pred > 0.5:
            z = 1
        else:
            z = 0
        
        if y[i] == z:
            correct +=1
    print(f"accuracy: {correct/len(y)}")
```

```python
accuracy(final_a ,  final_b, final_c, final_d, final_q,xs1,xs2,xs3,xs4,ys)
```

We get the accuracy which is almost equal to the accuracy obtained using the sklearn logistic regression.
